## 1. IMPORTING

```python
import tensorflow as tf
```

## 2. CONVERSION


```python
def convert_to_tflite(keras_model_path, tflite_output_path):
    model = tf.keras.models.load_model(keras_model_path) 
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    tflite_model = converter.convert()
    
    with open(tflite_output_path, "wb") as f:  
        f.write(tflite_model)
    
    print(f"Converted: {keras_model_path} → {tflite_output_path}")
    return tflite_output_path
convert_to_tflite(
    keras_model_path='Models/EfficientNetB3/efficientnetb3_model.keras',
    tflite_output_path='Models/tflite_conversion/best_model.tflite'
)
```
